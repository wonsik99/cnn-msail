# 2. More Indexing and Masking

```python
# Create a sample dataset of exam scores. Each row represents a "student"
scores = torch.tensor([[85, 92, 78, 96, 88],
                      [76, 89, 94, 82, 91],
                      [90, 87, 85, 93, 89],
                      [82, 78, 91, 86, 94]])

print("Scores shape:", scores.shape)
print("Scores:\n", scores)

# TODO: Find all scores above 90 and replace them with 90
# Your code here:
# capped_scores = scores.clone()
# capped_scores[capped_scores > 90] = 90
# print("Capped scores:\n", capped_scores)
new_scores = scores.clone()
new_scores[new_scores > 90] = 90
print(new_scores)

# TODO: Get the index of the maximum score in each row
# Your code here:
max_indices = torch.argmax(scores, dim=1)
print(max_indices)

# TODO: Create a boolean mask for students (rows) who have at least 3 scores above 85
# Your code here:
mask = (scores > 85).sum(dim=1) >= 3
print(mask)

# TODO: Calculate the average score for each student, but only for scores <= 90
# Hint: Use masking to exclude high scores -- use float('nan') to set values to nan -- then calculate mean
# Your code here:
new_scores = scores.clone().float()
new_scores[new_scores > 90] = float('nan')
avg_scores = torch.nanmean(new_scores, dim=1)
print(avg_scores)

# TODO: Get the scores of the top 2 performing students (based on sum of all scores)
# Hint: Check out the pytorch documentation for torch.topk()
# Your code here:
total_scores = scores.sum(dim=1)
top_values, top_indices = torch.topk(total_scores, k=2)
top_2 = scores[top_indices]
print(top_2)
```

### Solutions for Exercise 2 (attempt it on your own first)

```python
# Solutions for Exercise 2:
# Grade capping
capped_scores = scores.clone()
capped_scores[capped_scores > 90] = 90
print("Capped scores:\n", capped_scores)

# Index of maximum score in each row
max_indices = torch.argmax(scores, dim=1)
print("Max score indices per student:", max_indices)

# Boolean mask for students with at least 3 scores above 85
above_85 = scores > 85
students_with_3_above_85 = torch.sum(above_85, dim=1) >= 3
print("Students with 3+ scores above 85:", students_with_3_above_85)

# Average score excluding scores above 90
masked_scores = scores.clone().float()
masked_scores[scores > 90] = float('nan')
averages = torch.nanmean(masked_scores, dim=1)
print("Averages excluding >90:", averages)

# Top 2 performing students
total_scores = torch.sum(scores, dim=1)
top_2_indices = torch.topk(total_scores, 2).indices
top_2_scores = scores[top_2_indices]
print("Top 2 students' scores:\n", top_2_scores)
```
